```python
#Python Program to Implement Logical Operations over Images
#Programmer Name: Himanshu Sharma
#Program Date: 25/03/2021
#Program Time: 02:02PM
import numpy as np
import cv2 as cv


#This method reads the input image
def read_image(path_address):
    image = cv.imread(path_address)
    return image


#This method returns true/false based on if images have same dimensions or not
def check_for_same_shapes(image_1, image_2):
    image_1_shape = image_1.shape
    image_2_shape = image_2.shape
    same_shapes = image_1_shape == image_2_shape
    return same_shapes


#This method returns the type of input image is i.e. either color or grayscale
def get_image_type(image_1):
    shape = image_1.shape
    length = len(shape)
    return length


#This method performs logical and/nand on color images
def and_color(image_1, image_2, result, nand):
    row, column, space = image_1.shape
    if(nand):
        for i in range(row):
            for j in range(column):
                for k in range(space):
                    result[i,j,k] = not(image_1[i,j,k] and image_2[i,j,k])
    else:
        for i in range(row):
            for j in range(column):
                for k in range(space):
                    result[i,j,k] = image_1[i,j,k] and image_2[i,j,k]
    return result


#This method performs logical or/nor on color images
def or_color(image_1, image_2, result, nor):
    row, column, space = image_1.shape
    if(nor):
        for i in range(row):
            for j in range(column):
                for k in range(space):
                    result[i,j,k] = not(image_1[i,j,k] or image_2[i,j,k])
    else:
        for i in range(row):
            for j in range(column):
                for k in range(space):
                    result[i,j,k] = image_1[i,j,k] or image_2[i,j,k]
    return result


#This method performs logical xor/xnor on color images
def xor_color(image_1, image_2, result, xnor):
    row, column, space = image_1.shape
    if(xnor):
        for i in range(row):
            for j in range(column):
                for k in range(space):
                    temp1 = (not(image_1[i,j,k])) and image_2[i,j,k]
                    temp2 = image_1[i,j,k] and (not(image_2[i,j,k]))
                    result[i,j,k] = not( temp1 or temp2 )
    else:
        for i in range(row):
            for j in range(column):
                for k in range(space):
                    temp1 = (not(image_1[i,j,k])) and image_2[i,j,k]
                    temp2 = image_1[i,j,k] and (not(image_2[i,j,k]))
                    result[i,j,k] = temp1 or temp2
    return result


#This method performs logical not on color images
def not_color(image_1, result):
    row, column, space = image_1.shape
    for i in range(row):
        for j in range(column):
            for k in range(space):
                result[i,j,k] = not (image_1[i,j,k])
    return result


#This method performs logical and/nand on grayscale images
def and_(image_1, image_2, result, nand):
    row, column = image_1.shape
    if(nand):
        for i in range(row):
            for j in range(column):
                result[i,j] = not(image_1[i,j] and image_2[i,j])
    else:
        for i in range(row):
            for j in range(column):
                result[i,j] = image_1[i,j] and image_2[i,j]
    return result


#This method performs logical or/nor on grayscale images
def or_(image_1, image_2, result, nor):
    row, column = image_1.shape
    if(nor):
        for i in range(row):
            for j in range(column):
                result[i,j] = not(image_1[i,j] or image_2[i,j])
    else:
        for i in range(row):
            for j in range(column):
                result[i,j] = image_1[i,j] or image_2[i,j]
    return result


#This method performs logical xor/xnor on grayscale images
def xor_(image_1, image_2, result, xnor):
    row, column = image_1.shape
    if(xnor):
        for i in range(row):
            for j in range(column):
                temp1 = (not(image_1[i,j])) and image_2[i,j]
                temp2 = image_1[i,j] and (not(image_2[i,j]))
                result[i,j] = not( temp1 or temp2 )
    else:
        for i in range(row):
            for j in range(column):
                temp1 = (not(image_1[i,j])) and image_2[i,j]
                temp2 = image_1[i,j] and (not(image_2[i,j]))
                result[i,j] = temp1 or temp2
    return result


#This method performs logical not on grayscale images
def not_(image_1, result):
    row, column = image_1.shape
    for i in range(row):
        for j in range(column):
            result[i,j] = not (image_1[i,j])
    return result


#This method performs the required operation based on the type of given input
def perform_required_color_operation(image_1, image_2, result, operation):
    if(operation == "and"): result = and_color(image_1, image_2, result, False)
    if(operation == "or"): result = or_color(image_1, image_2, result, False)
    if(operation == "not"): result = not_color(image_1, result)
    if(operation == "nand"): result = and_color(image_1, image_2, result, True)
    if(operation == "nor"): result = or_color(image_1, image_2, result, True)
    if(operation == "xor"): result = xor_color(image_1, image_2, result, False)
    if(operation == "xnor"): result = xor_color(image_1, image_2, result, True)
    
    
#This method performs the required operation based on the type of given input
def perform_required_greyscale_operation(image_1, image_2, result, operation):
    if(operation == "and"): result = and_(image_1, image_2, result, False)
    if(operation == "or"): result = or_(image_1, image_2, result, False)
    if(operation == "not"): result = not_(image_1, result)
    if(operation == "nand"): result = and_(image_1, image_2, result, True)
    if(operation == "nor"): result = or_(image_1, image_2, result, True)
    if(operation == "xor"): result = xor_(image_1, image_2, result, False)
    if(operation == "xnor"): result = xor_(image_1, image_2, result, True)

    
#This method iterates over color image to perform desired operation
def iterate_over_color_image(image_1, image_2, operation):
    row,column,space = image_1.shape
    result = np.zeros(image_1.shape)
    perform_required_color_operation(image_1, image_2, result, operation)
    return result


#This method iterates over grayscale image to perform desired operation
def iterate_over_grayscale_image(image_1, image_2, operation):
    row,column = image_1.shape
    result = np.zeros(image_1.shape)
    perform_required_greyscale_operation(image_1, image_2, result, operation)
    return result


#This method produces the desired result of a logical operation
def iterate(image_1, image_2, operation):
    dimensions = get_image_type(image_1)
    #Based on the input dimensions, iterate
    if (dimensions == 2):
        result = iterate_over_grayscale_image(image_1, image_2, operation)
    if( dimensions == 3):
        result = iterate_over_color_image(image_1, image_2, operation)
    return result
    

#This method produces logical and operation result
def logical_and(image_1, image_2):
        same_shapes = check_for_same_shapes(image_1, image_2)
        result = np.zeros(image_1.shape)
        #If both images have different shape, addition is not possible
        if(not same_shapes):
            print("Difference in Shapes of Images(Logical And not Possible)")
        else:
            result = iterate(image_1, image_2, "and")
        return result

    
#This method produces logical nand operation result
def logical_nand(image_1, image_2):
        same_shapes = check_for_same_shapes(image_1, image_2)
        result = np.zeros(image_1.shape)
        #If both images have different shape, addition is not possible
        if(not same_shapes):
            print("Difference in Shapes of Images(Logical And not Possible)")
        else:
            result = iterate(image_1, image_2, "nand")
        return result    

    
#This method produces logical or operation result
def logical_or(image_1, image_2):
        same_shapes = check_for_same_shapes(image_1, image_2)
        result = np.zeros(image_1.shape)
        #If both images have different shape, addition is not possible
        if(not same_shapes):
            print("Difference in Shapes of Images(Logical And not Possible)")
        else:
            result = iterate(image_1, image_2, "or")
        return result

    
#This method produces logical nor operation result
def logical_nor(image_1, image_2):
        same_shapes = check_for_same_shapes(image_1, image_2)
        result = np.zeros(image_1.shape)
        #If both images have different shape, addition is not possible
        if(not same_shapes):
            print("Difference in Shapes of Images(Logical And not Possible)")
        else:
            result = iterate(image_1, image_2, "nor")
        return result    
    

#This method produces logical xor operation result
def logical_xor(image_1, image_2):
        same_shapes = check_for_same_shapes(image_1, image_2)
        result = np.zeros(image_1.shape)
        #If both images have different shape, addition is not possible
        if(not same_shapes):
            print("Difference in Shapes of Images(Logical And not Possible)")
        else:
            result = iterate(image_1, image_2, "xor")
        return result

    
#This method produces logical xnor operation result
def logical_xnor(image_1, image_2):
        same_shapes = check_for_same_shapes(image_1, image_2)
        result = np.zeros(image_1.shape)
        #If both images have different shape, addition is not possible
        if(not same_shapes):
            print("Difference in Shapes of Images(Logical And not Possible)")
        else:
            result = iterate(image_1, image_2, "xnor")
        return result    

#This method produces logical not operation result
def logical_not(image_1, image_2 = 0):
        result = np.zeros(image_1.shape)
        result = iterate(image_1, image_2, "not")
        return result    

#Main program starts here
    #Read images for the operations
circle = read_image(r"D:\Image Processing\Circle.png")
bars = read_image(r"D:\Image Processing\Bars.png")

    #Perform logical AND and NAND of images
and_result = logical_and(circle, bars)
nand_result = logical_nand(circle, bars)

    #Perform logical OR and NOR of images
or_result = logical_or(circle, bars)
nor_result = logical_nor(circle, bars)

    #Perform logical XOR and XNOR of images
xor_result = logical_xor(circle, bars)
xnor_result = logical_xnor(circle, bars)

    #Perform logical NOT on image
not_result = logical_not(circle)

    #Show original images
cv.imshow("Circle", circle)
cv.imshow("Bars", bars)
   
    #Show AND/NAND image result
cv.imshow("Logical AND", and_result)
cv.imshow("Logical NAND", nand_result)

    #Show OR/NOR image result
cv.imshow("Logical OR", or_result)
cv.imshow("Logical NOR", nor_result)

    #Show XOR/XNOR image result
cv.imshow("Logical XOR", xor_result)
cv.imshow("Logical XNOR", xnor_result)

    #Show NOT image result
cv.imshow("Logical NOT", not_result)

cv.waitKey(0)
cv.destroyAllWindows()
```
