```python
#Python Program to Implement Arithmetic Operations over Images
#Programmer Name: Himanshu Sharma
#Program Date: 23/03/2021
#Program Time: 01:53PM
import numpy as np
import cv2 as cv

#This method adds two images and return resultant images
def add_images(image_1, image_2):
    image_1_dimensions = image_1.shape
    image_2_dimensions = image_2.shape
    same_dimensions = image_1_dimensions == image_2_dimensions
    if not same_dimensions:
        print("The Dimensions of Images are not Same.\n")
        print("Addition is not possible.")
        return np.zeros(image_1_dimensions)
    else:
        added_image = np.add(image_1, image_2, dtype = np.uint8)
        return added_image

#This method adds a scalar value to any given image and return resultant image
def add_scalar(image_1, scalar):
    added_image = np.add(image_1, scalar, dtype = np.uint8)
    return added_image

#This method subtract two images and return resultant images
def subtract_images(image_1, image_2):
    image_1_dimensions = image_1.shape
    image_2_dimensions = image_2.shape
    same_dimensions = image_1_dimensions == image_2_dimensions
    if not same_dimensions:
        print("The Dimensions of Images are not Same.\n")
        print("Subtraction is not possible.")
        return np.zeros(image_1_dimensions)
    else:
        subtracted_image = np.subtract(image_1, image_2, dtype = np.uint8)
        return subtracted_image

#This method subtracts a scalar value to any given image and return resultant image 
def subtract_scalar(image_1, scalar):
    subtracted_image = np.subtract(image_1, scalar, dtype = np.uint8)
    return subtracted_image

#This method multiplies two images and return resultant images
def multiply_images(image_1, image_2):
    image_1_dimensions = image_1.shape
    image_2_dimensions = image_2.shape
    same_dimensions = image_1_dimensions == image_2_dimensions
    if not same_dimensions:
        print("The Dimensions of Images are not Same.\n")
        print("Multiplication is not possible.")
        return np.zeros(image_1_dimensions)
    else:
        multiplied_image = np.multiply(image_1, image_2, dtype = np.uint8)
        return multiplied_image

#This method multiplies a scalar value to any given image and return resultant images
def multiply_scalar(image_1, scalar):
    multiplied_image = np.multiply(image_1, scalar, dtype = np.uint8)
    return multiplied_image

#This method divides two images and return resultant images
def divide_images(image_1, image_2):
    image_1_dimensions = image_1.shape
    image_2_dimensions = image_2.shape
    same_dimensions = image_1_dimensions == image_2_dimensions
    divided_image = np.zeros(image_1_dimensions)
    
    if not same_dimensions:
        print("The Dimensions of Images are not Same.\n")
        print("Division is not possible.")
    else:
        row, column, space = image_1_dimensions
        for i in range (0, row, 1):
            for j in range(0, column, 1):
                for k in range(0, space, 1):
                    if(not (image_2[i, j, k] == 0)):
                        divided_image[i,j, k] = image_1[i,j, k]/image_2[i,j, k]
    return divided_image.astype(np.uint8)

#This method divides a scalar value to any given image and return resultant images
def divide_scalar(image_1, scalar):
    divided_image = np.divide(image_1,scalar)
    return divided_image.astype(np.uint8)

#Main program starts here
    
    #Reading input images
dog_image = cv.imread(r"D:\Image Processing\dog.jpg")
cat_image = cv.imread(r"D:\Image Processing\cat.jpeg")

    #Addition operation on images
added_image = add_images(dog_image, cat_image)
scalar_added_image = add_scalar(dog_image, 50)

    #Subtraction operation on images
subtracted_image = subtract_images(dog_image, cat_image)
scalar_subtracted_image = subtract_scalar(dog_image, 50)

    #Multiplication operaton on images
multiplied_image = multiply_images(dog_image, cat_image)
scalar_multiplied_image = multiply_scalar(dog_image, 2)

    #Division operaton on images
divided_image = divide_images(dog_image, cat_image)
scalar_divided_image = divide_scalar(dog_image, 2)

    #Showing the images of the addition results
cv.imshow("Added Image", added_image)
cv.imshow("Scalar Added Image", scalar_added_image)
    
    #Showing the images of the subtraction results
cv.imshow("Subtracted Image", subtracted_image)
cv.imshow("Scalar Subtracted Image", scalar_subtracted_image)
    
    #Showing the images of the multiplied results
cv.imshow("Multiplied Image", multiplied_image)
cv.imshow("Scalar Multiplied Image", scalar_multiplied_image)
    
    #Showing the images of the divided results
cv.imshow("Divided Image", divided_image)
cv.imshow("Scalar Divided Image", scalar_divided_image)

cv.waitKey(0)
cv.destroyAllWindows()
```
